### Calculate Mean

```python
def cal_mean(sample_data):
    return sum(sample_data) / len(sample_data)

mean = cal_mean(sample_data)
```

### Calculate Standard Deviation, Variance

```python
import math

def std(sample_data, mean):
    total = 0
    for data in sample_data:
        total += (data - mean) ** 2
        
    return (math.sqrt(total / (len(sample_data) - 1)), total / (len(sample_data) - 1))

sd, variance = std(sample_data, mean)
print(sd, variance)
```
