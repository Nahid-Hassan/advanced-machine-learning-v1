## Definition of the Derivative

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-20,20)
x
```

```python
def func(x):
    return 1 / (1 + x ** 2)
```

```python
plt.plot(x, func(x))
```

```python
def deri_func(x):
    return -2*x / ((1 + x**2) ** 2)
```

```python
plt.plot(x, deri_func(x))
```
