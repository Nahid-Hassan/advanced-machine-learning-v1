```python
# Scatter Plot
'''
DataFrame.plot.scatter(x, y, s=None, c=None, **kwargs)
    x = Horizontal Axis List/Series of Values 1-D array like
    y = Vertical
    c = Colormap
    s = Marker Size/Point Size
    **kwargs = df.plot(**kwargs) options or parameters
'''
import pandas as pd
```

```python
df = pd.DataFrame([[5.1, 3.5, 0], [4.9, 3.0, 0], [7.0, 3.2, 1],
                   [6.4, 3.2, 1], [5.9, 3.0, 2]],
                  columns=['length', 'width', 'species'])
```

```python
df.head()
```

```python
df.plot.scatter(x='length', y='width')
```

```python
df.plot.scatter(x='length', y='width', c='DarkBlue')
```

```python
df.plot.scatter(x='length', y='width', c='species', colormap='viridis')
```
