````python
from collections import Counter
````

````python
with open('datasets/id_lang_workwith.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguageWorkedWith'].split(';'))
print(language_counter)
````

````python
print(language_counter.most_common(15))
````

````python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
````

### Bar Plot

````python
# bar plot
plt.bar(languages, popularity)

# Decoration
plt.title('Most Popular Language in 2019')
plt.ylabel('Number of People who Use')

# plt.show()
````

#### Oh Messy in X-label

We also can use `barh()` function to solve this problem.

````python
# bar plot
plt.barh(languages, popularity)

# Decoration
plt.title('Most Popular Language in 2019')
plt.xlabel('Number of People who Use')

plt.show()
````

````python
# reverse
languages.reverse() # new: inplace reverse
popularity.reverse()

# bar plot
plt.barh(languages, popularity)

# Decoration
plt.title('Most Popular Language in 2019')
plt.xlabel('Number of People who Use')

plt.show()
````
