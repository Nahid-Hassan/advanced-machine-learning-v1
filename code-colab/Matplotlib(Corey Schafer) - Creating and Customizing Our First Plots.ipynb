### Import matplotlib

````python
from matplotlib import pyplot as plt
````

### Generate Dataset

````python
# value for x-axis
dev_x =  list(range(25, 36, 1))
dev_x
````

````python
# value for y-axis
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
dev_y
````

### Plot Dataset

    Simply use plt.plot(x, y) function.
    x(array-like/scalar): value for x-axis
    y(array-like/scalar): value for y-axis

### Draw a line plot

````python
# line plot
plt.plot(dev_x, dev_y) # generate
plt.show() # display our plot
````

### Add title, xlabel, ylabel

````python
# plot
plt.plot(dev_x, dev_y)

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

# show plot
plt.show()
````

### Another y-axis data

````python
from random import randint

"""
    randint(lower_bound, upper_bound)
"""

py_dev_y = [x + 2000 * randint(2,5)  for x in dev_y]
py_dev_y
````

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y)

# plot - 2
plt.plot(dev_x, py_dev_y)

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

# show plot
plt.show()
````

### Adding legend

#### Way-1: using plt.legend([.,.,.])

Now we get `two` lines. But we don't know which line represents which data points.

To overcome this prolem we can add `legends`.

```py
# .....

# put a list in order to ploting data
plt.legend(['All Devs', 'Py Devs']

# ....
```

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y)

plt.plot(dev_x, py_dev_y) # new: line plot

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')


plt.legend(['All Devs', 'Python Devs']) # new: adding legend

# show plot
plt.show()
````

#### Way-2: using label parameter with plt.plot(x, y, label='your legend label')

```py
# ....

plt.plot(dev_x, dev_y, label='All Devs')

# ...
```

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, label='All Devs') # new: adding labels

# plot - 2
plt.plot(dev_x, py_dev_y, label='Py Devs') # new: adding labels

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything

# show plot
plt.show()
````

### Add `format` string

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, 'k--', label='All Devs') # new: add format string

# plot - 2
plt.plot(dev_x, py_dev_y, 'b', label='Py Devs') # new: add format string

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything

# show plot
plt.show()
````

#### add `format` string more efficient way

```py
    fmt: [color][linestyle][marker]
    
    color = 'b' # b for blue
    linestyle = '--' # dash dash
    marker = '.' # dot style other like 'o' 

    `k--o`: [color][linestyle][marker]
        color: 'k'
        linestyle: '--'
        marker: 'o'
```

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, color='k', linestyle='--', marker='.', label='All Devs') # new: add format string

# plot - 2
plt.plot(dev_x, py_dev_y, color='b', marker='o', label='Py Devs') # new: add format string

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything

# show plot
plt.show()
````

#### Hex Colors

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, color='#000000', linestyle='--', marker='.', label='All Devs') # new: using hex colors

# plot - 2
plt.plot(dev_x, py_dev_y, color='#ff00ff', marker='o', label='Py Devs') # new: using hex colors

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything

# show plot
plt.show()
````

### Adding linewidth

```py
# .........

# linewidth = number value
plt.plot(..., linewidth=3,...)

# .......
```

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, color='#000000', linestyle='--', marker='.', label='All Devs')

# plot - 2
plt.plot(dev_x, py_dev_y, color='#ff0000', marker='o', linewidth=3, label='Py Devs') # new: add linewidth

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything

# show plot
plt.show()
````

#### Add `pading` and `grid`

```py
# .............

plt.tight_layout() # add padding
plt.grid(True) # add grid, default True

# ............
```

````python
# Multiline plot

# plot - 1
plt.plot(dev_x, dev_y, color='#000000', linestyle='--', marker='.', label='All Devs')

# plot - 2
plt.plot(dev_x, py_dev_y, color='#ff0000', marker='o', linewidth=3, label='Py Devs')

# add title/above the plot
plt.title('Median Salary (USD) by Age')
# adding x-axis label
plt.xlabel('Ages')
# adding y-axis label
plt.ylabel('Median Salary (USD)')

plt.legend() # new: don't need pass anything


# before plt.show()
plt.tight_layout() # new: add padding 
plt.grid(True) # new: add grid


# show plot
plt.show()
````
