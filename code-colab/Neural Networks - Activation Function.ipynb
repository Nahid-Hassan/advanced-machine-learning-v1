## Activation Function

### 1. SoftPlus Activation Function

![images](images/softplus-activation.jpg)

```python
from math import log, exp

def softplus(x):
    return log(1 + exp(x))
```

```python
# generate test set
import numpy as np

data = np.arange(-10, 10, .5)
data[:10]
```

```python
softplus_list = [softplus(x) for x in data]

# for i, x in zip(data, softplus_list):
#     print(i, '----', x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data, softplus_list)
```

### 2. Sigmoid Activation Function

![images](images/sigmoid.svg)

```python
def sigmoid(x):
    return 1 / (1 + exp(-x))

plt.plot(data, [sigmoid(x) for x in data])
```

### Relu and others

![images](images/all_activation_functions.png)

```python
def relu(x):
    return max(0, x)

plt.plot(data, list(map(relu, data)))
```

### Tanh

```python
from math import tanh

plt.plot(data, list(map(tanh, data)))
```

### Leaky Relu

```python
def leaky_relu(x):
    return max(.1 * x, x)

plt.plot(data, list(map(leaky_relu, data)))
```
